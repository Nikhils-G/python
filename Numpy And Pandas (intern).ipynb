```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.title("distribution of age:")
plt.hist(x=df["Age"],color="pink")
plt.xlabel("age")
plt.ylabel("frequency")
plt.show()
```

```python
sns.boxplot(x=df["Age"])
plt.title("hi")
plt.xlabel("ss")
plt.show()
```

```python
df["Sex"].value_counts().plot(kind="bar")
plt.title("sex distribution")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
df["Sex"].value_counts().plot(kind="barh")    #diddfernce is barh
plt.title("sex distribution")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show() 
```

```python
df["Pclass"].value_counts().plot(kind="barh",color="pink")
plt.title("Pclass distribution")
plt.xlabel("Pcalss")
plt.ylabel("Count")
plt.show()
```

```python
emb=df["Embarked"].value_counts()
keys=emb.keys().to_list()
counts=emb.to_list()
plt.pie(x=counts,labels=keys,autopct='%1.1f%%')
plt.show()
```
